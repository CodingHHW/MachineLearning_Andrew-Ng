# 基于Keras的 神经网络 识别手写数字

```python
from keras.datasets import mnist
import matplotlib.pyplot as plt

(X_train, y_train), (X_test, y_test) = mnist.load_data()
print type(X_train), X_train.shape
print type(y_train), y_train.shape
print type(X_test), X_test.shape
print type(y_test), y_test.shape
# print X_train[0]
# print y_train[0]
```

```python
plt.subplot(221)
plt.imshow(X_train[0])
plt.subplot(222)
plt.imshow(X_train[1])
plt.subplot(223)
plt.imshow(X_train[2])
plt.subplot(224)
plt.imshow(X_train[3])
plt.show()
```

## 使用神经网络，输入层784，隐藏层784，输出层10.用softmax多分类

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
from keras.utils import np_utils
# 数据预处理
n = X_train.shape[1] * X_train.shape[2]
X_train = X_train.reshape((X_train.shape[0], n)).astype('float32')
X_test = X_test.reshape((X_test.shape[0], n)).astype('float32')
# 讲X值归一化，不然准确率很低
X_train = X_train / 255
X_test = X_test / 255
# 标签独热处理
y_train = np_utils.to_categorical(y_train)
y_test = np_utils.to_categorical(y_test)
num_classes = y_test.shape[1]
print type(X_train), X_train.shape
print type(y_train), y_train.shape
print type(X_test), X_test.shape
print type(y_test), y_test.shape
# print X_train[0]
# print y_train[0]
```

```python
# 建立多层感知机（MLP）模型
model = Sequential()
model.add(Dense(n, input_dim=n)) # 输入层input_dim=784层，隐藏层也是n
model.add(Activation('relu')) # 激活函数使用relu函数，其他还有 sigmoid,tanh
model.add(Dropout(0.2)) # 防止过拟合
model.add(Dense(num_classes)) # 输出层num_classes=10层
model.add(Activation('softmax')) # 输出函数为softmax多元分类
model.compile(loss='categorical_crossentropy', metrics=["accuracy"], optimizer='adam') # 编译
model.fit(X_train, y_train, validation_data=(X_test, y_test), nb_epoch=10, batch_size=200, verbose=1)
# nb_epoch 为训练的轮数，nb统一表示为number of
# batch_size 为进行梯度下降时间每个批次包含的样本数
# verbose 为日志显示，0为不在标准输出流输出日志信息，1为输出进度条记录，2为每个epoch输出一行记录
# validation_data 为验证数据集，比较奇怪竟然将测试集当成了验证集
```

```python
score = model.evaluate(X_test, y_test, verbose=1)
```

```python
print score
```

```python
print('Test score:', score[0])
print('Test accuracy:', score[1]) # 正确率为98.25%
```

## 下面使用卷积神经网络 CNN

```python
import numpy as np
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers.convolutional import Convolution2D, MaxPooling2D # 卷积
from keras.utils import np_utils

(X_train, y_train), (X_test, y_test) = mnist.load_data()
X_train = X_train.reshape(X_train.shape[0], 1, 28, 28).astype('float32')
X_test = X_test.reshape(X_test.shape[0], 1, 28, 28).astype('float32')
X_train = X_train / 255
X_test = X_test / 255
y_train = np_utils.to_categorical(y_train)
y_test = np_utils.to_categorical(y_test)
num_classes = y_test.shape[1]
print type(X_train), X_train.shape
print type(y_train), y_train.shape
print type(X_test), X_test.shape
print type(y_test), y_test.shape
# 对于CNN，输入数据是三维张量，不能reshape为1维的，此数据格式应该与后端保持一致。
# print X_train[0]
```

```python
# 构建卷积神经网络SNN
model = Sequential()
# 32个5*5的卷积核，输入图像为1*28*28，input_shape在Theano中为(1,28,28)，在TensorFlow中为(28,28,1)
model.add(Convolution2D(32, 5, 5, border_mode='valid', input_shape=(1, 28, 28), activation='relu'))
# 降采样，就是用一个值来代替一块区域，这个值可以是区域的平均值，最大值，最小值等等，目的就是减少数据量。
# 使用MaxPooling，也就是取2*2的矩阵中最大的那一个代表这一块的特征。 经过降采样后，原来28*28的矩阵就只有14*14了，分成4块
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.2)) # 防止过拟合
model.add(Flatten()) # Flatten 将一个多维矩阵展开，铺平成一个一维的向量。为后面的全连接做准备。
# 从上面得到的向量应该就是32*4=128维
model.add(Dense(128, activation='relu')) # 这里是基本的神经网络，隐藏层128个，上面得到的1*128矩阵乘以128*num_classes得到结果
model.add(Dense(num_classes, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train, validation_data=(X_test, y_test), nb_epoch=10, batch_size=128, verbose=1)
```

```python
scores = model.evaluate(X_test, y_test, verbose=1)
```

```python
print('Test score:', scores[0])
print('Test accuracy:', scores[1]) # 正确率为98.93%
```

```python
hist = model.fit(X_train, y_train, validation_data=(X_test, y_test), nb_epoch=10, batch_size=128, verbose=1)
hist.history
```
